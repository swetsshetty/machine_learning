```python
## initially lets import numpy
import numpy as np
```

```python
l=[1,2,3,4,5]
a=np.array(l)
type(a)
print(a)

# to check the shape f array
print(a.shape) # specifies no of rows and columns: if 1 dimenentional-species the no. of elements
```

```python
#multi dimentioal array

l1=[1,2,7,4,0]
l2=[1,6,3,5,5]
l3=[4,2,3,4,5]

a1=np.array([l1,l2,l3])
print(a1)  # 2 dimentional array:2 [] barckets
print(a1.shape)

a1.reshape(1,15)
print(a1)
print(a1.shape) 
```

```python
# Indexing

## Accessing the array elements
# 1 dementional array
a=np.array([1,2,3,4,5,6,7,8])
print(a[3])

print(a1[:,:])
print(a1[:2,: 3])
print(a1[:3,: 3])
print(a1[:2,: 2])
print(a1[:2,2:4])
```

```python
# inbuilt functions

#arrange
a=np.arange(0,10)  # between 0-10 all the elements will be displayed
print(a)
a=np.arange(0,10, step=2)
print(a)

# line space
np.linspace(1,10,50) #  50 equi spaced values between 1-10

#braodcasting
a[3:]=100
print(a)
a1=a
a1[3:]=200
print(a1)
print(a)

#Copy function and braodcasting

a2=a.copy()
a2[3:]=500
print(a2)
print(a)

# ones
print(np.ones(4))
print(np.ones(4, dtype=int))
print(np.ones((2,5),dtype=int))



```

```python
# conditions useful in exploratory data analysis
a=np.array([1,2,3,4,5,6,7,8])
val=5
print(a<val)
print(a[a<val])

print(a*2)
print(a/2)
print(a%2)
print(a**2)
```

```python
# random distribution
np.random.rand(3,3)
np.random.randn(3,3)
np.random.randint(0,100,8).reshape(2,4)
np.random.random_sample((1,5))
```

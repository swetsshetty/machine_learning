```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
#correlation with heatmap: interlational between all the features: numerical data type


df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#joint plot : to univariant/ bi variant analysis
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
#joint plot : to uni variant analysis
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # drwa the best fit lone and PDF
```

```python
#pair plot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
#dist plot

sns.distplot(df["tips"],kde=False, bins=10)
```

```python
#CATEGORICAL PLOTS: 1) BOXPLOT 2) VIOLINPLOT 3) COUNTPLOT 4)BAR PLOT

sns.countplot('smoker', data=df)
```

```python
sns.countplot(y='smoker', data=df)
```

```python
#bar plot
sns.barplot(x='total_bill',y='smoker',data=df)
```
